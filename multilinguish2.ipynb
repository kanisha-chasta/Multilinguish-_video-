```python
# Import necessary libraries
import moviepy.editor as mp
import whisper
from transformers import pipeline
from googletrans import Translator
from gtts import gTTS
import os
```

```python
# Step 1: Upload video file
from google.colab import files
uploaded = files.upload()
video_path = list(uploaded.keys())[0]  # Get the uploaded video file name
video = mp.VideoFileClip(video_path)

# Step 2: Extract audio from video
audio_path = "audio.wav"
audio = video.audio
audio.write_audiofile(audio_path)
```

```python
# Step 3: Transcribe audio using Whisper
model = whisper.load_model("large")
result = model.transcribe(audio_path)
transcribed_text = result["text"]
print("Transcribed Text:", transcribed_text)

# Step 4: Detect emotion in text
emotion_classifier = pipeline("text-classification", model="bhadresh-savani/bert-base-uncased-emotion")
emotion_result = emotion_classifier(transcribed_text)
detected_emotion = emotion_result[0]['label']
print(f"Detected Emotion: {detected_emotion}")
```

```python
# Step 5: Translate text into target languages
translator = Translator()
target_languages = ["hi", "gu", "pa", "ta", "mr"]  # Hindi, Gujarati, Punjabi, Tamil, Marathi
translations = {}

for lang in target_languages:
    try:
        translations[lang] = translator.translate(transcribed_text, dest=lang).text
        print(f"Translation in {lang}: {translations[lang]}")
    except ValueError:
        print(f"Language '{lang}' is not supported by Google Translate.")
```

```python
# Step 6: Generate audio and lip-sync videos for each translation
for lang, translation in translations.items():
    # Adjust TTS parameters based on detected emotion
    slow = detected_emotion in ["sadness", "anger"]
    tts = gTTS(text=translation, lang=lang, slow=slow)

    # Save the audio file in .mp3 format and convert to .wav for lip-syncing
    mp3_audio_path = f"translated_audio_{lang}.mp3"
    wav_audio_path = f"translated_audio_{lang}.wav"
    tts.save(mp3_audio_path)
    os.system(f"ffmpeg -i {mp3_audio_path} -ar 16000 -ac 1 {wav_audio_path}")

    # Run Wav2Lip for lip-syncing
    output_video_path = f"dubbed_video_{lang}.mp4"
    !python Wav2Lip/inference.py --checkpoint_path Wav2Lip/checkpoints/wav2lip_gan.pth --face "{video_path}" --audio "{wav_audio_path}" --outfile "{output_video_path}"

print("Lip-synced videos created for each target language.")
```
